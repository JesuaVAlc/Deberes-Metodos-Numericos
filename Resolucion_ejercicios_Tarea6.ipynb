```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

Ejericio 1

```python
x_vals_plot = np.linspace(-0.25, 0.25, 400)
x_sym = sp.symbols('x')

# Funciones definidas
f1_expr = 1 / (25 * x_sym**2 + 1)
taylor_expr = 1 - 25 * x_sym**2 + 625 * x_sym**4
lagrange_expr = 1 - 22.5 * x_sym**2 + 250 * x_sym**4

# Convertir expresiones simbólicas a funciones numéricas
f1_func = sp.lambdify(x_sym, f1_expr, "numpy")
taylor_func = sp.lambdify(x_sym, taylor_expr, "numpy")
lagrange_func = sp.lambdify(x_sym, lagrange_expr, "numpy")

# Evaluar funciones
f1_vals = f1_func(x_vals_plot)
taylor_vals = taylor_func(x_vals_plot)
lagrange_vals = lagrange_func(x_vals_plot)
x_pts=[-0.2, -0.1, 0, 0.1, 0.2]
y_pts=[0.5, 0.8, 1, 0.8, 0.5]

# Graficar
plt.figure(figsize=(8, 5))
plt.plot(x_vals_plot, f1_vals, label="f(x) = 1/(25x² + 1)", linewidth=2)
plt.plot(x_vals_plot, taylor_vals, label="Taylor: 1 - 25x² + 625x⁴", linestyle='--')
plt.plot(x_vals_plot, lagrange_vals, label="Lagrange: 1 - 22.5x² + 250x⁴", linestyle='--')
plt.scatter(x_pts, y_pts, color='red', label='Puntos de Lagrange', zorder=3)

plt.title("Comparación entre f(x), Polinomio de Taylor y Lagrange")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

Ejercicio 2

```python
x_vals_plot = np.linspace(-5, 5, 400)
x_sym = sp.symbols('x')

# Función original y polinomios de aproximación
arctan_expr = sp.atan(x_sym)
taylor_expr = sp.pi/4 + (1/2)*(x_sym-1) - (1/4)*(x_sym-1)**2 + (1/12)*(x_sym-1)**3

# Polinomio de Lagrange obtenido previamente
lagrange_expr = 0.0064462*x_sym**4 + 0.04908186*x_sym**3 - 0.43624486*x_sym**2 + 1.19988*x_sym - 0.033773

# Puntos de interpolación
x_pts = [0.5, 0.75, 1.0, 1.25, 1.5]
y_pts = [np.arctan(x) for x in x_pts]

# Convertir expresiones simbólicas a funciones numéricas
arctan_func = sp.lambdify(x_sym, arctan_expr, "numpy")
taylor_func = sp.lambdify(x_sym, taylor_expr, "numpy")
lagrange_func = sp.lambdify(x_sym, lagrange_expr, "numpy")

# Evaluar funciones
arctan_vals = arctan_func(x_vals_plot)
taylor_vals = taylor_func(x_vals_plot)
lagrange_vals = lagrange_func(x_vals_plot)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_vals_plot, arctan_vals, label="f(x) = arctan(x)", linewidth=2, color='blue')
plt.plot(x_vals_plot, taylor_vals, label="Taylor (orden 4)", linestyle='--', color='green')
plt.plot(x_vals_plot, lagrange_vals, label="Lagrange (orden 4)", linestyle='-.', color='red')
plt.scatter(x_pts, y_pts, color='black', label='Puntos de interpolación', zorder=5)

plt.title("Comparación entre arctan(x) y sus aproximaciones polinómicas\n(centrado en x=1)", fontsize=12)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=10, loc='upper left')
plt.axvline(x=1, color='gray', linestyle=':', alpha=0.5)  # Línea vertical en x=1

# Ajustar límites para mejor visualización
plt.xlim(0.4, 1.6)
plt.ylim(0.4, 1.1)

plt.tight_layout()
plt.show()
```
