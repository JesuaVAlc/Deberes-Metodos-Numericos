**1)** Dados los puntos (0,1), (1,5), (2,3), determine el spline c√∫bico. 

\begin{align*}
S_0(x)&= a_0+b_0(x-x_0)+c_0(x-x_0)^{2}+d_0(x-x_0)^{3}\\
&= a_0+b_0x+c_0x^{2}+d_0x^{3}\\
S_1(x)&=a_1+b_1(x-x_1)+c_0(x-x_1)^{2}+d_1(x-x_1)^{3}\\
&=a_1+b_1(x-1)+c_0(x-1)^{2}+d_0(x-1)^{3}
\end{align*}

**[Ecuacion 1]**

\begin{gather*}
S_0(x_0) = y_0 \Rightarrow a_0+b_0x+c_0x^{2}+d_0x^{3} \\
a_0=y_0 \Rightarrow a_0=1\\
\end{gather*}

**[Ecuacion 2]**

\begin{gather*}
S_0(x_1) = y_1 \Rightarrow 1+b_0x+c_0x^{2}+d_0x^{3} \\
b_0x+c_0x^{2}+d_0x^{3}= 4
\end{gather*}

**[Ecuacion 3]**

\begin{gather*}
S_1(x_1) = y_1 \Rightarrow a_1+b_1(x-1)+c_0(x-1)^{2}+d_1(x-1)^{3} \\
a_0=y_1 \Rightarrow a_1=5
\end{gather*}

**[Ecuacion 4]**

\begin{gather*}
S_1(x_2) = y_2 \Rightarrow 5+b_1(x-1)+c_0(x-1)^{2}+d_1(x-1)^{3} \\
b_1+c_0+d_1 = -2\\
\end{gather*}

**[Ecuacion 5]**

$S'_i = b_i+2c_i(x-x_i)+3d_i(x-x_i)^{2}$

\begin{align*}
S'_0(x_1) &= S'_1(x_1)\\
b_0+2c_0x+3d_0x^{2} &= b_1+2c_1(x-1)^{2}+3d_1(x-1)^{2}\\
b_0+2c_0+3d_0^{2} &= b_1\\
\end{align*}

**[Ecuacion 6]**

$S''_i = 2c_i+6d_i(x-x_i)$

\begin{align*}
S''_0(x_1) &= S''_1(x_1)\\
2c_0+6d_0x &= 2c_1+6d_1(x-1)\\
2c_0+6d_0 &= 2c_1\\
\end{align*}

**[Ecuacion 7] Frontera Natural**

\begin{gather*}
S''_0(x_0) = 0 \\
2c_0=0 \\
c_0=0\\
\end{gather*}


**[Ecuacion 7] Frontera Natural**

\begin{gather*}
S''_1(x_2) = 0 \\
2c_1+6d_1(x_2-x_1)=0\\
2c_1+6d_1=0\\
\end{gather*}

**Sistema de ecuaciones**
\begin{align*}
\begin{cases}
b_0+d_0=4\\
b_1+c_1+d_1=-2\\
b_0+3d_0=b_1\\
3d_0=c_1\\
c_1+3d_1=0\\
\end{cases}
\end{align*}

**Splines Completos**

\begin{align*}
\begin{cases}
S_0(x)= 1+ \frac{11}{2}x-\frac{3}{2}x^{3}\\
S_1(x)= 5+ (x-1) -\frac{9}{2}(x-1)^{2} -\frac{3}{2}(x-1)^{3}
\end{cases}
\end{align*}

````python
import numpy as np
import matplotlib.pyplot as plt

# Puntos dados
x_points = np.array([0, 1, 2])
y_points = np.array([1, 5, 3])

# Definici√≥n de los splines
def S0(x):
    return 1 + (11/2)*x - (3/2)*x**3

def S1(x):
    return 5 + (x-1) - (9/2)*(x-1)**2 + (3/2)*(x-1)**3

# Generaci√≥n de puntos para graficar
x0 = np.linspace(0, 1, 100)
x1 = np.linspace(1, 2, 100)

# Gr√°fico
plt.figure(figsize=(8, 5))
plt.scatter(x_points, y_points, color='red', label='Puntos originales', zorder=5)
plt.plot(x0, S0(x0), label='$S_0(x)$', color='blue')
plt.plot(x1, S1(x1), label='$S_1(x)$', color='green')
plt.title('Spline C√∫bico Natural')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
````

````python
import sympy as sym
from IPython.display import display


# ###################################################################
def cubic_spline(xs: list[float], ys: list[float], return_intervals=False):
    """
    Cubic spline interpolation ``S``. Every two points are interpolated by a cubic polynomial
    ``S_j`` of the form ``S_j(x) = a_j + b_j(x - x_j) + c_j(x - x_j)^2 + d_j(x - x_j)^3.``

    xs must be different  but not necessarily ordered nor equally spaced.

    ## Parameters
    - xs, ys: points to be interpolated

    ## Return
    - List of symbolic expressions for the cubic spline interpolation.
    """

    points = sorted(zip(xs, ys), key=lambda x: x[0])  # sort points by x

    xs = [x for x, _ in points]
    ys = [y for _, y in points]

    n = len(points) - 1  # number of splines

    h = [xs[i + 1] - xs[i] for i in range(n)]  # distances between  contiguous xs

    alpha = [0] * n
    for i in range(1, n):
        alpha[i] = 3 / h[i] * (ys[i + 1] - ys[i]) - 3 / h[i - 1] * (ys[i] - ys[i - 1])

    l = [1]
    u = [0]
    z = [0]

    for i in range(1, n):
        l += [2 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1]]
        u += [h[i] / l[i]]
        z += [(alpha[i] - h[i - 1] * z[i - 1]) / l[i]]

    l.append(1)
    z.append(0)
    c = [0] * (n + 1)

    x = sym.Symbol("x")
    splines = []
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d = (c[j + 1] - c[j]) / (3 * h[j])
        a = ys[j]
        print(j, a, b, c[j], d)
        S = a + b * (x - xs[j]) + c[j] * (x - xs[j]) ** 2 + d * (x - xs[j]) ** 3

        splines.append(S)
    splines.reverse()
    if return_intervals:
        intervals = [(xs[j], xs[j+1]) for j in range(n)]
        return splines, intervals
    return splines
````

````python
 ###################################################################
def cubic_spline_clamped(
    xs: list[float], ys: list[float], B0: float, B1: float
) -> list[sym.Symbol]:
    """
    Cubic spline interpolation ``S``. Every two points are interpolated by a cubic polynomial
    ``S_j`` of the form ``S_j(x) = a_j + b_j(x - x_j) + c_j(x - x_j)^2 + d_j(x - x_j)^3.``

    xs must be different  but not necessarily ordered nor equally spaced.

    ## Parameters
    - xs, ys: points to be interpolated
    - B0, B1: derivatives at the first and last points

    ## Return
    - List of symbolic expressions for the cubic spline interpolation.
    """

    points = sorted(zip(xs, ys), key=lambda x: x[0])  # sort points by x
    xs = [x for x, _ in points]
    ys = [y for _, y in points]
    n = len(points) - 1  # number of splines
    h = [xs[i + 1] - xs[i] for i in range(n)]  # distances between  contiguous xs

    alpha = [0] * (n + 1)  # prealloc
    alpha[0] = 3 / h[0] * (ys[1] - ys[0]) - 3 * B0
    alpha[-1] = 3 * B1 - 3 / h[n - 1] * (ys[n] - ys[n - 1])

    for i in range(1, n):
        alpha[i] = 3 / h[i] * (ys[i + 1] - ys[i]) - 3 / h[i - 1] * (ys[i] - ys[i - 1])

    l = [2 * h[0]]
    u = [0.5]
    z = [alpha[0] / l[0]]

    for i in range(1, n):
        l += [2 * (xs[i + 1] - xs[i - 1]) - h[i - 1] * u[i - 1]]
        u += [h[i] / l[i]]
        z += [(alpha[i] - h[i - 1] * z[i - 1]) / l[i]]

    l.append(h[n - 1] * (2 - u[n - 1]))
    z.append((alpha[n] - h[n - 1] * z[n - 1]) / l[n])
    c = [0] * (n + 1)  # prealloc
    c[-1] = z[-1]

    x = sym.Symbol("x")
    splines = []
    for j in range(n - 1, -1, -1):
        c[j] = z[j] - u[j] * c[j + 1]
        b = (ys[j + 1] - ys[j]) / h[j] - h[j] * (c[j + 1] + 2 * c[j]) / 3
        d = (c[j + 1] - c[j]) / (3 * h[j])
        a = ys[j]
 
        S = a + b * (x - xs[j]) + c[j] * (x - xs[j]) ** 2 + d * (x - xs[j]) ** 3

        splines.append(S)
    splines.reverse()
    return splines
````

````python
xs3 = [27.7, 28, 29, 30]
ys3 = [4.1, 4.3, 4.1, 3.0]
B0_3 = 0.33  # f'(x0)
B1_3 = -1.5  # f'(xn)
````

````python
from sympy.utilities.lambdify import lambdify
````

````python
plt.figure(figsize=(12, 8))

x_sym = sym.Symbol('x')
splines1 = cubic_spline_clamped(xs1, ys1, B0_1, B1_1)

# Graficar cada segmento de la curva 1
for i in range(len(splines1)):
    x_start, x_end = xs1[i], xs1[i+1]
    x_vals = np.linspace(x_start, x_end, 100)
    
    # Convertir expresi√≥n simb√≥lica a funci√≥n num√©rica
    f = lambdify(x_sym, splines1[i], 'numpy')
    y_vals = f(x_vals)
    
    plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='Curva 1' if i == 0 else "")

splines2 = cubic_spline_clamped(xs2, ys2, B0_2, B1_2)

# Graficar cada segmento de la curva 2
for i in range(len(splines2)):
    x_start, x_end = xs2[i], xs2[i+1]
    x_vals = np.linspace(x_start, x_end, 100)
    
    f = lambdify(x_sym, splines2[i], 'numpy')
    y_vals = f(x_vals)
    
    plt.plot(x_vals, y_vals, 'r-', linewidth=2, label='Curva 2' if i == 0 else "")

splines3 = cubic_spline_clamped(xs3, ys3, B0_3, B1_3)

# Graficar cada segmento de la curva 3
for i in range(len(splines3)):
    x_start, x_end = xs3[i], xs3[i+1]
    x_vals = np.linspace(x_start, x_end, 100)
    
    f = lambdify(x_sym, splines3[i], 'numpy')
    y_vals = f(x_vals)
    
    plt.plot(x_vals, y_vals, 'g-', linewidth=2, label='Curva 3' if i == 0 else "")

# Graficar todos los puntos
all_xs = xs1 + xs2[1:] + xs3[1:]
all_ys = ys1 + ys2[1:] + ys3[1:]
plt.scatter(all_xs, all_ys, color='black', s=80, zorder=5, label='Puntos de datos')

# Destacar puntos de uni√≥n entre curvas
plt.scatter([17, 27.7], [4.5, 4.1], color='magenta', s=100, zorder=6, 
            edgecolors='black', label='Puntos de uni√≥n')

# Configuraci√≥n del gr√°fico
plt.title('Splines C√∫bicos Sujetos para las Tres Curvas', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()

plt.show()
````
